### Working with matplotlib

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1, 3, 10, 12],[1, 6, 11, 13])
plt.title('Samarth')
plt.xlabel('time[s]')
plt.ylabel('speed[m]')
plt.show()
```

```python
from matplotlib import style



style.use('ggplot')

x = [0,2, 3]
y = [0,4, 6]
x2 = [0, 5, 9]
y2 = [0, 2, 10]

plt.plot(x, y, label='first', linewidth=2)
plt.plot(x2, y2, label='second', linewidth=6)
plt.legend()
plt.show()
```

```python
histogram_x = [5, 5, 5, 25, 100, 90, 50]
plt.hist(histogram_x)
plt.show()
```
